```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../datasets/emails.csv')
df.head(10)
```

```python
df.shape
```

```python
df.info()
```

```python
# Dropping uncessary columns
df.drop(columns=['Email No.'],axis=1, inplace=True)
df.head(10)
```

```python
df.isnull().sum().sum()
```

```python
# Checking if data is baised
display(df['Prediction'].value_counts()[0])
display(df['Prediction'].value_counts()[1])
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

**KNN Classifier**

```python
# Elbow method for finding optimal number of clusters

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

wcss = []
for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    wcss.append(accuracy_score(y_test, y_pred))
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,10), wcss)
plt.title('Elbow Curve')
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train, y_train)
```

```python
y_pred = knn.predict(X_test)
```

```python
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

**Support Vector Machines**

```python
from sklearn.svm import SVC
```

```python
svm_linear = SVC(kernel='linear',C=1)
svm_linear.fit(X_train, y_train)
y_pred = svm_linear.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
svm_rbf = SVC(kernel='rbf', C=1)
svm_rbf.fit(X_train,y_train)
y_pred = svm_rbf.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```

```python
svm_poly = SVC(kernel='poly',C=1)
svm_poly.fit(X_train,y_train)
y_pred = svm_poly.predict(X_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred))
```
