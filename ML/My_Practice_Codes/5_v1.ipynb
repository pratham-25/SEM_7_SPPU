## K Means Clustering

```python
# Removing Outliers as K Means is sensitive to outliers
for col in dfk.iloc[:,0:11].columns:
    plt.figure(figsize=(6,6))
    sns.boxplot(x=dfk[col])
    plt.title(col)
```

```python
def clean_outliers(data, cols):
    for col in cols:
        q1 = np.percentile(data[col], 25)
        q3 = np.percentile(data[col], 75)
        iqr = q3-q1
        upper = q3 + 1.5*iqr
        lower = q1 - 1.5*iqr
        
        data = data[(data[col] > lower) & (data[col] < upper)]
        
    return data
```

```python
outlier_colums = ['QUANTITYORDERED','MSRP','SALES']
dfk = clean_outliers(data=dfk, cols=outlier_colums)
```

```python
for col in dfk.iloc[:,0:11].columns:
    plt.figure(figsize=(6,6))
    sns.boxplot(x=dfk[col])
    plt.title(col)
```

```python
dfk.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dfk_scaled = scaler.fit_transform(dfk)
dfk_scaled
```

**Elbow Method**

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.metrics import silhouette_score
wcss = []
silhoutte = []

for i in range(2,20):
    kmeans = KMeans(n_clusters=i)
    labels = kmeans.fit_predict(dfk_scaled)
    wcss.append(kmeans.inertia_)
    silhoutte.append(silhouette_score(dfk_scaled,labels))
    
```

```python
# Elbow Curve
plt.figure(figsize=(10,8))
plt.plot(range(2,20), wcss, marker='o')
```

```python
# Silhoutte Curve
plt.figure(figsize=(10,8))
plt.plot(range(2,20), silhoutte, marker='o')
```

```python
kmeans = KMeans(n_clusters=10, init='k-means++', random_state=2)
labels = kmeans.fit_predict(dfk_scaled)
labels
```
