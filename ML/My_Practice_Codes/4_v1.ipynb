```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def objective(x):
    return (x+3)**2
```

```python
def derivative(x):
    return 2*(x+3)
```

```python
def gradientDescent(start, lr, epoch):
    x_list = []
    x_old = start
    for i in range(epoch):
        slope = derivative(x_old)
        x_new = x_old - (slope * lr)
        x_list.append(x_new)
        x_old = x_new
    return x_list
```

```python
lr = 0.1
epoch = 100
start = 2
```

```python
x_list = gradientDescent(start, lr, epoch)
x_list
```

```python
from sympy import Derivative
from sympy.abc import x
```

```python
df = Derivative((x+3)**2,x, evaluate=True)
df
```

```python
print(df.doit().subs(x,x_list[0]))
print(df.doit().subs(x,x_list[-1]))
```
